# Telco Customer Churn Analysis 📊

## 1. Introduction
- Business problem: Predict why customers leave
- Dataset: ~7,000 customers from a telco company

## 2. Data Cleaning
- Dropped missing/empty values
- Converted `TotalCharges` to float
- Encoded categorical features

## 3. Exploratory Data Analysis (EDA)
- Churn distribution
- Churn by contract type, tenure, monthly charges
- Stacked bar charts, churn rates, SHAP summary













## 4. Modeling
- Models used: Logistic Regression, Random Forest, XGBoost
- Performance metrics: Accuracy, Precision, Recall, F1, AUC
- Tuning: RandomizedSearchCV for XGBoost

## 5. Model Comparison
| Model               | Accuracy | Recall | AUC  |
|--------------------|----------|--------|------|
| Logistic Regression| 0.78     | 0.50   | 0.80 |
| Random Forest       | 0.80     | 0.47  | 0.83 |
| XGBoost (tuned)     | 0.8      | 0.51   | 0.85 |

## 7. Final Thoughts & Recommendations
- Retain long-term customers via 1–2 year contracts
- High churn in fiber optic users → investigate service quality
- Payment method affects churn (electronic check = risk)
